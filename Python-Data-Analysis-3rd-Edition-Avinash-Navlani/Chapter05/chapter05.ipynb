### Accessories for Charts

In the **Matplotlib** module, we can enhance the readability and interpretability of graphs by adding titles, axis labels, legends, and more. These accessories help in better understanding the data and its representation.

#### Adding Titles and Labels
- **Title**: A title can be added to the graph using the `plt.title()` function.
- **Axis Labels**: Labels for the x-axis and y-axis can be added using `plt.xlabel()` and `plt.ylabel()` respectively.

#### Multiple Graphs on a Single Plot
- Matplotlib allows plotting multiple graphs (e.g., line, bar, scatter) on a single figure.
- Points from different series can be displayed on the same graph.

#### Legends
- A **legend** is a box that appears on the graph, typically on the right or left side, and explains what each element of the graph represents.
- Legends are particularly useful when multiple series are plotted on the same graph.
- Legends can be added using the `plt.legend()` function.

#### Example
Below is an example demonstrating how to use these accessories in a chart:

```python
#Add the required libraries
import matplotlib.pyplot as plt

# Create the data
x = [1,3,5,7,9,11]
y = [10,25,35,33,41,59]

# Let's plot the data
plt.plot(x, y,label='Series-1', color='blue')

# Create the data
x = [2,4,6,8,10,12]
y = [15,29,32,33,38,55]

# Plot the data
plt.plot(x, y, label='Series-2', color='red')

# Add X Label on X-axis
plt.xlabel("X-label")

# Add X Label on X-axis
plt.ylabel("Y-label")

# Append the title to graph
plt.title("Multiple Python Line Graph")

# Add legend to graph
plt.legend()

# Display the plot
plt.show()
```

### Scatter plot
Scatter plots draw data points using Cartesian coordinates to show the values of numerical
values. They also represent the relationship between two numerial values. We can create a
scatter plot in Matplotlib using the scatter() function, as follows:

```python
# Add the essential library matplotlib
import matplotlib.pyplot as plt

# create the data
x = [1,3,5,7,9,11]
y = [10,25,35,33,41,59]
# Draw the scatter chart
plt.scatter(x, y, c='blue', marker='*',alpha=0.5)
# Append the label on X-axis
plt.xlabel("X-label")

# Append the label on X-axis
plt.ylabel("Y-label")
# Add the title to graph
plt.title("Scatter Chart Sample")
# Display the chart
plt.show()
```

### Line plot

A line plot is a chart that displays a line between two variables. It has a sequence of data
points joined by a segment:

```python
# Add the essential library matplotlib

import matplotlib.pyplot as plt

# create the data
x = [1,3,5,7,9,11]
y = [10,25,35,33,41,59]

# Draw the line chart
plt.plot(x, y)

# Append the label on X-axis
plt.xlabel("X-label")

# Append the label on X-axis
plt.ylabel("Y-label")

# Append the title to chart
plt.title("Line Chart Sample")

# Display the chart
plt.show()
```

### Pie plot
A pie plot is a circular graph that is split up into wedge-shaped pieces. Each piece is
proportionate to the value it represents. The total value of the pie is 100 percent:

```python
# Add the essential library matplotlib

import matplotlib.pyplot as plt
# create the data
subjects = ["Mathematics","Science","Communication Skills","Computer Application"]
scores = [85,62,57,92]

# Plot the pie plot
plt.pie(scores,
labels=subjects,
colors=['r','g','b','y'],
startangle=90,
shadow= True,
explode=(0,0.1,0,0),
autopct='%1.1f%%')
# Add title to graph
plt.title("Student Performance")
# Draw the chart
plt.show()
```

### Bar plot
A bar plot is a visual tool to compare the values of various groups. It can be drawn horizontally or vertically. We can create a bar graph using the bar() function:

```python
# Add the essential library matplotlib
import matplotlib.pyplot as plt
# create the data
movie_ratings = [1,2,3,4,5]
rating_counts = [21,45,72,89,42]
# Plot the data
plt.bar(movie_ratings, rating_counts, color='orange')
# Add X Label on X-axis
plt.xlabel("Movie Ratings")
# Add X Label on X-axis
plt.ylabel("Rating Frequency")
# Add a title to graph
plt.title("Movie Rating Distribution")
# Show the plot
plt.show()
```

### Histogram plot
A histogram shows the distribution of a numeric variable. We create a histogram using the hist() method. It shows the probability distribution of a continuous variable. A histogram only works on a single variable while a bar graph works on two variables:

```python
# Add the essential library
import matplotlib.pyplot as plt
# Create the data
employee_age = [21,28,32,34,35,35,37,42,47,55]
# Create bins for histogram
bins = [20,30,40,50,60]
# Plot the histogram

plt.hist(employee_age, bins, rwidth=0.6)
# Add X Label on X-axis
plt.xlabel("Employee Age")
# Add X Label on X-axis
plt.ylabel("Frequency")
# Add title to graph
plt.title("Employee Age Distribution")
# Show the plot
plt.show()
```

### Bubble plot
A bubble plot is a type of scatter plot. It not only draws data points using Cartesian coordinates but also creates bubbles on data points. Bubble shows the third dimension of a plot. It shows three numerical values: two values are on the x and y axes and the third one
is the size of data points (or bubbles):

```python
# Import the required modules
import matplotlib.pyplot as plt
import numpy as np

# Set figure size
plt.figure(figsize=(8,5))

# Create the data
countries =['Qatar','Luxembourg','Singapore','Brunei','Ireland','Norway','UAE','Kuwait']
populations = [2781682,
604245,5757499,428963,4818690,5337962,9630959,4137312]
gdp_per_capita = [130475, 106705, 100345, 79530, 78785, 74356,69382, 67000]

# scale GDP per capita income to shoot the bubbles in the graph
scaled_gdp_per_capita = np.divide(gdp_per_capita, 80)
colors = np.random.rand(8)


# Draw the scatter diagram
plt.scatter(countries, populations, s=scaled_gdp_per_capita, c=colors,
cmap="Blues",edgecolors="grey", alpha=0.5)

# Add X Label on X-axis
plt.xlabel("Countries")

# Add Y Label on X-axis
plt.ylabel("Population")

# Add title to graph
plt.title("Bubble Chart")

# rotate x label for clear visualization
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

### Pandas Plotting

The **pandas** library provides a convenient `plot()` method, which acts as a wrapper around the **Matplotlib** library. This method allows users to create various types of plots directly from pandas DataFrames or Series.

#### Key Parameters of the `plot()` Method:
1. **`kind`**:
    - Specifies the type of graph to create.
    - Common options include:
      - `line`: Line plot (default).
      - `bar`: Vertical bar plot.
      - `barh`: Horizontal bar plot.
      - `hist`: Histogram.
      - `box`: Box plot.
      - `kde`: Kernel Density Estimate plot.
      - `pie`: Pie chart.
      - `area`: Area plot.
      - `scatter`: Scatter plot.

2. **`figsize`**:
    - Defines the size of the figure as a tuple `(width, height)`.

3. **`title`**:
    - Sets the title of the graph.

4. **`grid`**:
    - A boolean parameter to enable or disable grid lines on the axis.

5. **`legend`**:
    - Controls the display of the legend.

6. **`xticks`**:
    - Specifies the sequence of ticks on the x-axis.

7. **`yticks`**:
    - Specifies the sequence of ticks on the y-axis.

#### Advantages of Pandas Plotting:
- Simplifies the process of visualizing data directly from DataFrames or Series.
- Reduces the need for extensive Matplotlib code.
- Provides quick and easy visualizations for exploratory data analysis.

```python
# Import the required modules

import pandas as pd
import matplotlib.pyplot as plt

# Let's create a Dataframe
df = pd.DataFrame({
'name':['Ajay','Malala','Abhijeet','Yming','Desilva','Lisa'],
'age':[22,72,25,19,42,38],
'gender':['M','F','M','M','M','F'],
'country':['India','Pakistan','Bangladesh','China','Srilanka','UK'],
'income':[2500,3800,3300,2100,4500,5500]
})

# Create a scatter plot
df.plot(kind='scatter', x='age', y='income', color='red', title='Age Vs Income')

# Show figure
plt.show()
```

In the preceding plot, the plot() function takes kind, x, y, color, and title values. In our example, we are plotting the scatter plot between age and income using the kind parameter as 'scatter'. The age and income columns are assigned to the x and y parameters. The scatter point color and the title of the plot are assigned to the color and title parameters:

```python
import matplotlib.pyplot as plt
import pandas as pd
#Create bar plot

# Create bar plot
df.plot(kind='bar',x='name', y='age', color='orange')
# Show figure
plt.show()
```

### lm plots
The lm plot plots the scatter and fits the regression model on it. A scatter plot is the best
way to understand the relationship between two variables. Its output visualization is a joint
distribution of two variables. lmplot() takes two column names – x and y – as a string and
DataFrame variable.

```python
# Import the required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create DataFrame
df=pd.DataFrame({'x':[1,3,5,7,9,11],'y':[10,25,35,33,41,59]})

# Create lmplot
sns.lmplot(x='x', y='y', data=df)

# Show figure
plt.show()
```

```python
# Load the dataset
df=pd.read_csv("HR_comma_sep.csv")
# Create lmplot
sns.lmplot(x='satisfaction_level', y='last_evaluation', data=df,
fit_reg=False, hue='left')
# Show figure
plt.show()
```

In the above example, last_evaluation is the evaluated performance of the employee, satisfaction_level is the employee's satisfaction level in the company, and left means whether the employee left the company or not. satisfaction_level and last_evaluation were drawn on the x and y axes, respectively. The third variable left is passed in the hue parameter. The hue property is used for color shade. We are passing a left variable as hue. We can clearly see in the diagram that employees that have left are scattered into three groups.

### Bar Plots

The `barplot()` function in Seaborn provides a way to visualize the relationship between a **categorical variable** and a **continuous variable**. It uses rectangular bars where the length of each bar represents the value of the continuous variable.

### Box Plots

A **box plot**, also known as a **box-whisker plot**, is a powerful visualization tool used to understand the distribution of a variable, including its quartiles, minimum, maximum, and outliers.

#### Key Features of Box Plots:
1. **Quartiles**:
    - The box represents the interquartile range (IQR), which is the range between the first quartile (Q1) and the third quartile (Q3).
    - The line inside the box indicates the median (second quartile, Q2).

2. **Whiskers**:
    - The whiskers extend from the box to the minimum and maximum values within a specified range (typically 1.5 times the IQR).

3. **Outliers**:
    - Data points outside the whiskers are considered outliers and are plotted as individual points.

4. **Orientation**:
    - Box plots can be drawn horizontally or vertically, depending on the context.

#### Advantages of Box Plots:
- Provides a clear summary of the distribution of a variable.
- Highlights the presence of outliers in the data.
- Useful for comparing distributions across multiple groups.

Box plots are widely used in exploratory data analysis to gain insights into the spread and variability of data.

```python
# Import the required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a new DataFrame with the required columns
df_new = pd.DataFrame({
	'satisfaction_level': [0.5, 0.7, 0.8, 0.6, 0.9],
	'last_evaluation': [0.6, 0.8, 0.7, 0.9, 0.85]
})

# Create boxplot
sns.boxplot(data=df_new[['satisfaction_level', 'last_evaluation']])
# Show figure
plt.show()
```

### KDE Plots

The **KDE (Kernel Density Estimation)** plot is a statistical tool used to estimate the probability density function of a continuous variable. It provides a smooth curve that represents the distribution of the data.

#### Key Features of KDE Plots:
1. **Non-Parametric Estimation**:
    - KDE is a non-parametric method, meaning it does not assume any specific distribution for the data.

2. **Smooth Curve**:
    - The plot provides a smooth representation of the data's distribution, making it easier to interpret compared to histograms.

3. **Kernel Function**:
    - The kernel function determines the shape of the curve. Common kernels include Gaussian, Epanechnikov, and Tophat.

4. **Bandwidth**:
    - Bandwidth controls the smoothness of the curve. A smaller bandwidth results in a more detailed curve, while a larger bandwidth produces a smoother curve.

#### Advantages of KDE Plots:
- Provides a clear visualization of the data's distribution.
- Useful for identifying patterns, peaks, and multimodal distributions.

#### Example:
The `kde()` function in Seaborn can be used to create KDE plots for a given continuous variable.

```python

# Import the required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Create density plot
sns.kdeplot(df_new.satisfaction_level)
# Show figure
plt.show()
```

### Violin Plots

A **violin plot** is a powerful visualization tool that combines the features of a **box plot** and a **KDE (Kernel Density Estimation)** plot. It provides a detailed representation of the distribution of a dataset, making it easier to analyze and interpret.

#### Key Features of Violin Plots:
1. **Combination of Box Plot and KDE**:
    - The central part of the violin plot resembles a box plot, showing the interquartile range (IQR), median, and outliers.
    - The outer shape represents the KDE, providing a smooth estimate of the data's probability density.

2. **Symmetry**:
    - The KDE is mirrored on both sides of the central axis, creating a symmetrical shape that resembles a violin.

3. **Distribution Insights**:
    - The width of the violin at different points indicates the density of the data at those values.
    - Wider sections represent higher density, while narrower sections represent lower density.

4. **Multiple Groups**:
    - Violin plots are particularly useful for comparing the distributions of multiple groups or categories.

5. **Customization**:
    - Violin plots can be customized to include or exclude elements like the box plot, KDE, or individual data points.

#### Advantages of Violin Plots:
- Combines the summary statistics of a box plot with the detailed distribution insights of a KDE plot.
- Provides a clear visualization of multimodal distributions (distributions with multiple peaks).
- Useful for comparing distributions across multiple groups or categories.

#### Use Cases:
- Analyzing the distribution of numerical data across different categories.
- Identifying patterns, outliers, and variations in the data.
- Comparing the spread and density of data across multiple groups.

#### Example:
Below is an example of a violin plot using the Seaborn library:

```python
# Create violin plot
sns.violinplot(data=df_new[['satisfaction_level', 'last_evaluation']])
# Show figure
plt.show()
```

### Heatmaps

A **heatmap** is a two-dimensional graphical representation of data where individual values contained in a matrix are represented as colors. It is a powerful visualization tool for identifying patterns, correlations, and trends in data.

#### Key Features of Heatmaps:
1. **Grid Representation**:
    - The data is displayed in a grid format, where each cell represents a value from the matrix.

2. **Color Encoding**:
    - The values in the matrix are encoded as colors, with a color gradient indicating the magnitude of the values.

3. **Annotations**:
    - Heatmaps can include annotations to display the exact value of each cell within the grid.

4. **Customizable Color Palettes**:
    - Heatmaps allow the use of various color palettes to represent data effectively.

5. **Correlation Analysis**:
    - Heatmaps are commonly used to visualize correlation matrices, making it easier to identify relationships between variables.

#### Advantages of Heatmaps:
- Provides a clear and concise representation of complex data.
- Highlights patterns, clusters, and outliers in the data.
- Useful for exploratory data analysis and feature selection.

#### Use Cases:
- Visualizing correlation matrices in statistical analysis.
- Representing data distributions in machine learning.
- Analyzing patterns in time-series data.
- Displaying hierarchical clustering results.

#### Example:
Below is an example of creating a heatmap using the Seaborn library in Python:

```python
# Import required library
import seaborn as sns
# Read iris data using load_dataset() function
data = sns.load_dataset("iris")
# Exclude non-numerical columns
numerical_data = data.select_dtypes(include=['float64', 'int64'])
# Find correlation
cor_matrix = numerical_data.corr()
# Create heatmap
sns.heatmap(cor_matrix, annot=True)
# Show figure
plt.show()
```

The Iris dataset is loaded using the `load_dataset()` function, and the correlation matrix is calculated using the `corr()` method. The correlation matrix is then visualized using the `heatmap()` function, which provides a grid view of the matrix. The `annot` parameter is set to `True` to display the correlation values within the grid. The resulting plot is a symmetric matrix where the diagonal values are all ones, indicating a perfect correlation of each variable with itself. Additionally, a custom color map can be applied using the `cmap` parameter to enhance the visualization.

```python
# Create heatmap
sns.heatmap(cor_matrix, annot=True, cmap="YlGnBu")
# Show figure
plt.show()
```

### Layouts in Bokeh

Bokeh provides various layout options for organizing plots and widgets in a single panel. These layouts are essential for creating interactive visualizations and dashboards. They also support resizing plots and widgets based on the panel size by setting sizing modes.

#### Types of Layouts:
1. **Row Layout**:
    - Organizes all the plots in a row or horizontal fashion.
    - Useful for displaying multiple plots side by side.

2. **Column Layout**:
    - Organizes all the plots in a column or vertical fashion.
    - Suitable for stacking plots or widgets vertically.

3. **Nested Layout**:
    - Combines row and column layouts.
    - Allows for more complex arrangements of plots and widgets.

4. **Grid Layout**:
    - Arranges plots in a grid or matrix format.
    - Ideal for displaying multiple plots in a structured grid.

By leveraging these layout options, you can create visually appealing and well-organized dashboards for data visualization.

### Multiple Plots Using Grid Layout

#### Overview
A **grid layout** is a powerful feature in Bokeh that allows users to arrange multiple plots and widget objects in a row-column matrix format. This layout is particularly useful for creating dashboards and complex visualizations where multiple plots need to be displayed together.

#### Key Features of Grid Layout:
1. **Row-Column Matrix**:
    - Plots and widgets are arranged in a structured grid format, making it easy to organize and compare visualizations.

2. **Flexible Placement**:
    - The grid layout accepts a list of figure objects for each row.
    - `None` can be used as a placeholder to leave empty spaces in the grid.

3. **Customizable**:
    - The size and arrangement of the grid can be customized to suit the requirements of the visualization.

#### Advantages of Grid Layout:
- Provides a clean and organized way to display multiple plots.
- Enhances the readability and interpretability of visualizations.
- Useful for creating dashboards and multi-plot layouts.

#### Example:
Below is an example of how to use a grid layout in Bokeh to arrange multiple plots:

```python
# Import the required modules
from bokeh.plotting import figure
from bokeh.plotting import output_notebook
from bokeh.plotting import show
from bokeh.layouts import gridplot
# Import iris flower dataset as pandas DataFrame
from bokeh.sampledata.iris import flowers as df
# Output to notebook
output_notebook()
# Instantiate a figure
fig1 = figure(width=300, height=300)
fig2 = figure(width=300, height=300)
fig3 = figure(width=300, height=300)
# Create scatter marker plot by render the circles
fig1.circle(df['petal_length'], df['sepal_length'], size=8, color =
"green", alpha = 0.5)
fig2.circle(df['petal_length'], df['sepal_width'], size=8, color = "blue",
alpha = 0.5)
fig3.circle(df['petal_length'], df['petal_width'], size=8, color = "red",
alpha = 0.5)
# Create a grid layout
grid_layout = gridplot([[fig1, fig2], [None,fig3]])
# Show the plot
show(grid_layout)
```

### Hide click policy
Hide click policy hides the desirable glyphs by clicking on the legend entry. Let's see an
example of a hide click policy

```python
# Import the required modules
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource
from bokeh.sampledata.iris import flowers as df

# Output to notebook
output_notebook()

# Instantiate a figure object
fig = figure(
	width=500,
	height=350,
	title="Petal length Vs. Petal Width",
	x_axis_label='petal_length',
	y_axis_label='petal_width'
)

# Create scatter marker plot by rendering the circles
for specie, color in zip(['setosa', 'virginica', 'versicolor'], ['blue', 'green', 'red']):
	data = df[df['species'] == specie]
	source = ColumnDataSource(data)
	fig.circle(
		'petal_length',
		'petal_width',
		size=8,
		color=color,
		alpha=0.7,
		legend_label=specie,
		source=source
	)

# Set the legend location and click policy
fig.legend.location = 'top_left'
fig.legend.click_policy = "hide"

# Show the plot
show(fig)
```
